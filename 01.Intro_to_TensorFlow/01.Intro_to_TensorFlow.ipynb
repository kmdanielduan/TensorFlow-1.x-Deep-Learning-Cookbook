## 1.3 Hello World

```
# import tensorflow
import tensorflow as tf
```

```
# tf.constant
message = tf.constant('Hello World!')
```

```
# define tf.Session() as sess and run the Session
with tf.Session() as sess:
  print(sess.run(message).decode())
  
# sess.run(message) gets b'Hello World!'
```
